### KMean Method Elbow Clustering

```python
X = df[['Type','C1','C2','C3','C4','C5','C6']]
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km = KMeans(n_clusters=i,n_init=20)
    km.fit(X)
    css = km.inertia_
    print('Clusters=',i," wcss= ",css)
    wcss.append(css)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("No of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
km = KMeans(n_clusters=3,n_init=20)
```

```python
km.fit(X)
```

```python
cluster_labels = km.predict(X)
```

```python
cluster_labels
```

```python
plt.scatter(X['Type'],X['C1'],c = cluster_labels)
plt.xlabel('Type')
plt.ylabel('C1')
plt.show()
```
